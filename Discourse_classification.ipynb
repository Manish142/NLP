## Removing the punctuations from the text data

```python
#data = df_train.iloc[:,2]
#x_train['2'].replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
import re
corpous = []
for i in range(len(x_train['2'])):
    review = re.sub('[^a-zA-Z]', ' ', x_train['2'][i])
    review = review.lower()
    corpous.append(review)
#print(corpous)
#x_train
```

Text vectorisation

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector=CountVectorizer(max_features = 5000, ngram_range=(2,2))
Vec_corpous=countvector.fit_transform(corpous)
```
